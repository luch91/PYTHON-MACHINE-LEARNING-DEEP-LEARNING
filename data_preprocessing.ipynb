Outlier Treatment by capping the value

```python
# identify the 99 and 1 percentile
np.percentile(df.n_hot_rooms,[99])
```

```python
upper_value = np.percentile(df.n_hot_rooms,[99])[0]
upper_value
```

```python
# identify the rows where the numbers are greater than the upper_limit
df[(df.n_hot_rooms >upper_value)]
```

```python
# capping the value
df.n_hot_rooms[df.n_hot_rooms >3 * upper_value]
```
